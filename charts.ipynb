```python
#2. plt.plot method can accept two lists for arguments
# The first one corresponds to X-Axis and the second one corresponds to the Y-Axis(Y-Coordinate)

X=['Num1','Num2','Num3','Num4','Num5','Num6']
Y=np.arange(6)
plt.plot(X,Y, marker= 's')
```

```python
#3. Plots in matplotlib resides withine a figure of object.Therefore, You need to create a new figure and add subplots to each figure.
fig=plt.figure()
# 2X2 Matrix
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
# 3X3 Matrix 2,2 should be 3,3 in all axes
#ax5=fig.add_subplot(2,2,5)
#ax6=fig.add_subplot(2,2,6)
#ax7=fig.add_subplot(2,2,7)
```

```python
#4. different line styles
plt.plot(np.random.randn(100).cumsum(),'r') #solid line
plt.plot(np.random.randn(100).cumsum(),'r--') #dotted line
```

```python
plt.plot(np.random.randn(100).cumsum(),'r-') #solid line
```

```python
plt.plot(np.random.randn(100).cumsum(),'k') #solid line by default
```

```python
#5. Another way to get figures and subplot
fig, ax=plt.subplots(1,2)
ax[0].plot(np.random.rand(100).cumsum(),'r--')
ax[1].plot(np.random.rand(100).cumsum(),'b')
```

```python
fig, ax=plt.subplots(2,1)
ax[0].plot(np.random.rand(100).cumsum(),'r--')
ax[1].plot(np.random.rand(100).cumsum(),'b')
```

```python
#6 3D, Subplot
fig, ax=plt.subplots(3,3)
ax[0,0].plot(np.random.randn(100).cumsum(),'r--')
ax[0,1].hist(np.random.randn(100),bins=10,color='b')
ax[1,0].scatter(np.random.randn(100),np.random.randn(100))
```

```python
#7. Legend creation
data=np.random.randn(50).cumsum()
plt.plot(data,'b--', label='default')
plt.plot(data,'b--', drawstyle='steps-post',label='steps-post')
plt.legend(loc='best')
```

```python
# with location set as 'center'
data=np.random.randn(80).cumsum()
plt.plot(data,'b--', label='default')
plt.plot(data,'b--', drawstyle='steps-post',label='steps-post')
plt.legend(loc='center') 
```

```python
# chart with title , Xlable & Ylable
fig=plt.figure()
ax=fig.add_subplot(3,3,3)
ax.plot(np.random.randn(1000).cumsum(), label='random cumulative sum')
ax.set_xticks([0,250,500,750,1000])
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), label='random cumulative sum')
```

```python
fig=plt.figure()
ax=fig.add_subplot(2,2,2)
ax.plot(np.random.randn(1000).cumsum(), label='random cumulative sum')
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), label='Random cumsum')
ax.set_xticks([0,250,500,750,1000])
ax.set_title('My Plot')
ax.set_xlabel('Stages')
ax.set_ylabel('Values')
ax.legend(loc='best')
```
