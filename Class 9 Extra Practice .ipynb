You need to know when to use pivot_table and when to use crosstab with two keys. Both pivot_table and crosstab are similar to grouped by with two keys. Both returns a table(values in one key are used as rows and values in another key are used as columns) or you can think of it as another DataFrame. By using pivot_table, you can use any aggregation function applied to the grouped by structure such as mean(), max(), min() or size(). However, using crosstab. it automatically return the values by using the size() function which means how many happening. Please truly understand the difference of pivot_table and crosstab. Crosstab is one case of pivot_table.

By using pivot_table, you start from a reference or varibale of a DataFrame. By using crosstab, You start with the pandas library. Please practice the following:

1. Create a pivot table of average price based on year and bedrooms_count
2. Show the number of listing based on year and bedrooms_count

```python
pd.crosstab(airbnb['date_of_listing'].dt.year, airbnb['bedrooms_count'])
```
