```python
airbnb =pd.read_csv('airbnb.csv', index_col=0)
airbnb.dropna(subset=['average_rate_per_night', 'bedrooms_count','date_of_listing', 'latitude', 'longitude' ], inplace =True)
airbnb.head()
```

```python
airbnb['average_rate_per_night']=airbnb['average_rate_per_night'].str.replace('$','')
airbnb['average_rate_per_night']=airbnb['average_rate_per_night'].astype(int)
airbnb['date_of_listing']=pd.to_datetime(airbnb['date_of_listing'])
airbnb.head()
```

```python
_2bd=airbnb[airbnb['bedrooms_count']=='2']
_2bd['average_rate_per_night'].mean()
```

```python
grouped_2bd =_2bd.groupby(airbnb['date_of_listing'].dt.year)['average_rate_per_night'].mean()
grouped_2bd
```

```python
ax =grouped_2bd.plot(marker='s')
ax.set_title('2BD Room Price by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Price')
```

```python
ax =grouped_2bd.plot.bar()
ax.set_title('2BD Room Price by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Price')
```

```python
pivot_year_bd=airbnb.pivot_table('average_rate_per_night', index='bedrooms_count', columns=airbnb['date_of_listing'].dt.year,aggfunc='mean')
pivot_year_bd
```

```python
pivot_year_bd.T.drop(2008)[['1','2','3']].plot(marker='s')
```

```python
air=airbnb[airbnb['city']=='Houston']
cross=pd.crosstab(air['date_of_listing'].dt.year, air['bedrooms_count'])
cross
```

```python
cross.plot.barh(stacked= True)
```

```python
Cities = airbnb[airbnb['city'].isin(['Houston', 'Dallas','Austin', 'San Antonio'])]

houstonData = Cities[Cities['city'] == 'Houston']
dallasData =Cities[Cities['city'] == 'Dallas']
austinData =Cities[Cities['city'] == 'Austin']
sanAntoData = Cities[Cities['city'] == 'San Antonio']
```

```python
houston_cTab = pd.crosstab(houstonData['date_of_listing'].dt.year, houstonData['bedrooms_count'])
dallas_cTab = pd.crosstab(dallasData['date_of_listing'].dt.year, dallasData['bedrooms_count'])
austin_cTab = pd.crosstab(austinData['date_of_listing'].dt.year, austinData['bedrooms_count'])
sanAnto_cTab = pd.crosstab(sanAntoData['date_of_listing'].dt.year, sanAntoData['bedrooms_count'])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
houston_cTab.plot.barh(stacked= True, ax=axes[0,0])
dallas_cTab.plot.barh(stacked= True, ax=axes[0,1])
austin_cTab.plot.barh(stacked= True, ax=axes[1,0])
sanAnto_cTab.plot.barh(stacked= True, ax=axes[1,1])
axes[0,0].set_title('Number of Listings of Houston')
axes[0,0].set_ylabel('Year')
axes[0,1].set_title('Number of Listings of Dallas')
axes[0,1].set_ylabel('Year')
axes[1,0].set_title('Number of Listings of Ausitn')
axes[1,0].set_ylabel('Year')
axes[1,1].set_title('Number of Listings of San Antonio')
axes[1,1].set_ylabel('Year')
```
